```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
feautres_na = [features for features in df.columns if df[features].isnull().sum()>1]
```

```python
feautres_na
```

```python
for features in feautres_na:
    print(features,np.round(df[features].isnull().mean(),4),'%')
```

```python
more_than_50_percent_misssing_value_features = [features for features in feautres_na if (np.round(df[features].isnull().mean(),4)) > 0.5]
```

```python
more_than_50_percent_misssing_value_features
```

```python
for feature in feautres_na:
    data = df.copy()
    data[feature] = np.where(data[feature].isnull(),1,0) # replacing missing value with 1 and rest with 0
    data.groupby(feature)['SalePrice'].mean().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
numerical_features = [features for features in df.columns if df[features].dtypes != 'O']
numerical_features
```

```python
df[numerical_features].head()
```

```python
year_features = [features for features in numerical_features if "Yr" in features or "Year" in features]
year_features
```

```python
df.groupby('YrSold')['SalePrice'].median().plot()
```

```python
for feature in year_features:
    if feature != "YrSold":
        data = df.copy()
        data[feature] = data['YrSold'] - data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.title(feature)
        plt.show()
```

```python
discrete_features = [feature for feature in numerical_features if len(df[feature].unique())<25 and feature not in year_features]
discrete_features
```

```python
df[discrete_features].head()
```

```python
for feature in discrete_features:
    data = df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
continous_features = [feature for feature in numerical_features if feature not in discrete_features+year_features+['Id']]
continous_features
```

```python
for feature in continous_features:
    data = df.copy()
    data[feature].hist(bins = 25)
    plt.title(feature)
    
    plt.show()
```

```python
for feature in continous_features:
    data = df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

```python
for feature in continous_features:
    data = df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        #plt.scatter(data[feature],data['SalePrice'])
        data[feature].hist(bins = 25)
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

```python
categorical_features_test =[feature for feature in df_test.columns if df_test[feature].dtypes == 'O']
categorical_features_test
```

```python
for feature in categorical_features:
    print("Distinct categories for {}  are {}".format(feature,len(df[feature].unique())))
```

```python
for feature in categorical_features:
    data = df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
for feature in categorical_features:
    df[feature] = df[feature].fillna(df[feature].mode()[0])                 #train
    
```

```python
for feature in categorical_features_test:
    df_test[feature] = df_test[feature].fillna(df_test[feature].mode()[0]) 
```

```python
for feature in discrete_features:
    df[feature] = df[feature].fillna(df[feature].mode()[0])                 #train
    df_test[feature] = df_test[feature].fillna(df_test[feature].mode()[0])                #train
    
```

```python
for feature in year_features:
    df[feature] = df[feature].fillna(df[feature].mode()[0])                 #train
    df_test[feature] = df_test[feature].fillna(df_test[feature].mode()[0]) 
```

```python
continous_numerical_features = [feature for feature in continous_features if feature not in 'SalePrice']
```

```python
for feature in continous_numerical_features:
    df[feature] = df[feature].fillna(df[feature].mean())                 #train
    df_test[feature] = df_test[feature].fillna(df_test[feature].mean())
```

```python
for feature in more_than_50_percent_misssing_value_features:
    df.drop([feature],axis = 1, inplace = True)
    df_test.drop([feature],axis = 1, inplace = True)
```

```python
df.drop(['Id'],axis = 1, inplace = True)
```

```python
df_test.drop(['Id'],axis = 1, inplace = True)
```

```python
x=x.drop(['Street','Utilities','Condition2','Heating','Electrical','GarageCond','MiscVal'],axis=1)
```

```python
df_test=df_test.drop(['Street','Utilities','Condition2','Heating','Electrical','GarageCond','MiscVal'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20, random_state=0)
```

```python
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor , ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
xgb = XGBRegressor()
rfr = RandomForestRegressor()
dt = DecisionTreeRegressor()
```

```python
print(lr.fit(x_train , y_train))
print(xgb.fit(x_train , y_train))
print(rfr.fit(x_train , y_train))
print(dt.fit(x_train , y_train))
```

```python
lr_y_pred=lr.predict(x_test)
xgb_y_pred=xgb.predict(x_test)
rfr_y_pred=rfr.predict(x_test)
dt_y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
print(lr.score(x_train,y_train))
print(xgb.score(x_train,y_train))
print(rfr.score(x_train,y_train))
print(dt.score(x_train,y_train))
```

```python
print(lr.score(x_test,y_test))
print(xgb.score(x_test,y_test))
print(rfr.score(x_test,y_test))
print(dt.score(x_test,y_test))
```

```python
print(r2_score(lr.predict(x_train) , y_train))
print(r2_score(xgb.predict(x_train) , y_train))
print(r2_score(rfr.predict(x_train) , y_train))
print(r2_score(dt.predict(x_train) , y_train))
```

```python
print(r2_score(lr_y_pred , y_test))
print(r2_score(xgb_y_pred , y_test))
print(r2_score(rfr_y_pred , y_test))
print(r2_score(dt_y_pred , y_test))
```

```python
print('MAE:', mean_absolute_error(y_test, lr_y_pred))
print('MAE:', mean_absolute_error(y_test, xgb_y_pred))
print('MAE:', mean_absolute_error(y_test, rfr_y_pred))
print('MAE:', mean_absolute_error(y_test, dt_y_pred))
```

```python
print('RMSE:', np.sqrt(mean_squared_error(y_test, lr_y_pred)))
print('RMSE:', np.sqrt(mean_squared_error(y_test, xgb_y_pred)))
print('RMSE:', np.sqrt(mean_squared_error(y_test, rfr_y_pred)))
print('RMSE:', np.sqrt(mean_squared_error(y_test, dt_y_pred)))
```
